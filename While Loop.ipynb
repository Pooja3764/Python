```python
#Print first 10 natural numbers
i=1
while i<=10:
  print(i)
  i=i+1
```

```python
#Sum of first 10 natural nmbr
i=1
sum=0
while i<=10:
  sum=sum+i
  i=i+1
print("The sum is",sum)    # Note necessary to keep print just below the while otherwise
                              # if we keep below i then it will be considered as a part of loop and will print every time.
```

```python
# Table of a given number.
n=int(input("enter the numbr"))
i=1
while i<=10:
  print(n*i)
  i=i+1
```

```python
# Table of a given number.
n=int(input("enter the numbr"))
i=1
while i<=10:
  print(n,"*",i,"=",n*i)
  i=i+1
```

```python
# Factorial
n=int(input("enter the number"))
f=1
while n>0:
  f=f*n
  n=n-1
print("the factorial is",f)
```

```python
# Factors of a given number.
n=int(input("enter the number "))
i=1
while i<=n:
  if n%i==0:
    print(i,end=" ")
  i=i+1
```

```python
#sum of even and odd numbers.
i=1
se=0
so=0
while i<=100:
  if i%2==0:
     se=se+i
  else:
     so=so+i
  i=i+1
print("The sum of even numbers is", se)
print("The sum of odd numbers is", so)
```

```python
# the sum of digits of a number

n=int(input("enter the number"))
sum=0
while n>0:
  rem=n%10
  sum=sum+rem
  n=n//10
print("the sum of digits is", sum)
```

```python
#reverse of a number
n=int(input("enter the number"))
rev=0
while n>0:
  rem=n%10
  rev=rev*10+rem
  n=n//10
print("the reverse is", rev)
```
